```python
import pandas as pd
import numpy as np

# Реальные средние температуры (примерные данные) для городов по сезонам
seasonal_temperatures = {
    "New York": {"winter": 0, "spring": 10, "summer": 25, "autumn": 15},
    "London": {"winter": 5, "spring": 11, "summer": 18, "autumn": 12},
    "Paris": {"winter": 4, "spring": 12, "summer": 20, "autumn": 13},
    "Tokyo": {"winter": 6, "spring": 15, "summer": 27, "autumn": 18},
    "Moscow": {"winter": -10, "spring": 5, "summer": 18, "autumn": 8},
    "Sydney": {"winter": 12, "spring": 18, "summer": 25, "autumn": 20},
    "Berlin": {"winter": 0, "spring": 10, "summer": 20, "autumn": 11},
    "Beijing": {"winter": -2, "spring": 13, "summer": 27, "autumn": 16},
    "Rio de Janeiro": {"winter": 20, "spring": 25, "summer": 30, "autumn": 25},
    "Dubai": {"winter": 20, "spring": 30, "summer": 40, "autumn": 30},
    "Los Angeles": {"winter": 15, "spring": 18, "summer": 25, "autumn": 20},
    "Singapore": {"winter": 27, "spring": 28, "summer": 28, "autumn": 27},
    "Mumbai": {"winter": 25, "spring": 30, "summer": 35, "autumn": 30},
    "Cairo": {"winter": 15, "spring": 25, "summer": 35, "autumn": 25},
    "Mexico City": {"winter": 12, "spring": 18, "summer": 20, "autumn": 15},
}

# Сопоставление месяцев с сезонами
month_to_season = {12: "winter", 1: "winter", 2: "winter",
                   3: "spring", 4: "spring", 5: "spring",
                   6: "summer", 7: "summer", 8: "summer",
                   9: "autumn", 10: "autumn", 11: "autumn"}

# Генерация данных о температуре
def generate_realistic_temperature_data(cities, num_years=10):
    dates = pd.date_range(start="2010-01-01", periods=365 * num_years, freq="D")
    data = []

    for city in cities:
        for date in dates:
            season = month_to_season[date.month]
            mean_temp = seasonal_temperatures[city][season]
            # Добавляем случайное отклонение
            temperature = np.random.normal(loc=mean_temp, scale=5)
            data.append({"city": city, "timestamp": date, "temperature": temperature})

    data = pd.DataFrame(data)
    data['season'] = data['timestamp'].dt.month.map(lambda x: month_to_season[x])
    return data

# Генерация данных
data = generate_realistic_temperature_data(list(seasonal_temperatures.keys()))
data.to_csv('temperature_data.csv', index=False)
```

# 1. Анализ временных рядов

```python
# Группируем данные по городу и сортируем по дате
data = data.sort_values(by=['city', 'timestamp'])
data
```

```python
data['rolling_avg_30d'] = data.groupby('city')['temperature'].transform(lambda x: x.rolling(window=30, min_periods=1).mean())
data
```

```python
import matplotlib.pyplot as plt

city_data = data[data['city'] == 'Beijing']

# Визуализация
plt.figure(figsize=(10, 6))
plt.plot(city_data['timestamp'], city_data['temperature'], label='Daily Temperature', marker='o')
plt.plot(city_data['timestamp'], city_data['rolling_avg_30d'], label='30-Day Rolling Average', linestyle='--')
plt.title('Temperature and 30-Day Rolling Average for CityA', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Temperature', fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
seasonal_stats = data.groupby(['city', 'season'])['temperature'].agg(['mean', 'std']).reset_index()
seasonal_stats.rename(columns={'mean': 'avg_temperature', 'std': 'std_temperature'}, inplace=True)

seasonal_stats
```

```python
seasonal_stats.to_csv('seasonal_stats.csv')
```

```python
data = data.merge(seasonal_stats[['city', 'season', 'avg_temperature']],
                         on=['city', 'season'],
                         how='left')
data
```

```python
data = data.merge(seasonal_stats[['city', 'season', 'std_temperature']],
                         on=['city', 'season'],
                         how='left')
data
```

```python
# Определяем границы для аномалий: ±2σ
data['upper_bound'] = data['rolling_avg_30d'] + 2 * data['std_temperature']
data['lower_bound'] = data['rolling_avg_30d'] - 2 * data['std_temperature']

# Определяем аномалии
data['is_anomaly'] = (data['temperature'] > data['upper_bound']) | (data['temperature'] < data['lower_bound'])

# Фильтруем данные для города "Beijing"
city_data = data[data['city'] == 'Beijing']

# Визуализация
plt.figure(figsize=(10, 6))
plt.plot(city_data['timestamp'], city_data['temperature'], label='Daily Temperature', marker='o')
plt.plot(city_data['timestamp'], city_data['rolling_avg_30d'], label='30-Day Rolling Average', linestyle='--')
plt.fill_between(city_data['timestamp'], city_data['lower_bound'], city_data['upper_bound'], color='gray', alpha=0.2, label='±2σ Range')
plt.scatter(city_data['timestamp'][city_data['is_anomaly']],
            city_data['temperature'][city_data['is_anomaly']],
            color='red', label='Anomalies', zorder=5)
plt.title('Temperature Anomalies for CityA', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Temperature', fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Отображение таблицы аномалий для анализа
anomalies = city_data[city_data['is_anomaly']]
anomalies
```

# 2. Мониторинг текущей температуры

```python
API_KEY = "ee783dd5c25ec4dcb5a47afa9c85cb20"
WEATHER_URL = "http://api.openweathermap.org/data/2.5/weather"
```

```python
import requests
import asyncio
import aiohttp

def get_current_temperature(city):
    params = {
        "q": city,
        "appid": API_KEY,
        "units": "metric",
    }
    response = requests.get(WEATHER_URL, params=params)
    data = response.json()
    return data["main"]["temp"]
```

```python
def is_anomaly(cur_temp, city, season):
  cur_norm_temp = seasonal_stats[(seasonal_stats['city'] == city) & (seasonal_stats['season'] == season)]
  avg_temp, std_temp = cur_norm_temp['avg_temperature'].values[0], cur_norm_temp['std_temperature'].values[0]
  return (cur_temp < avg_temp - 2 * std_temp) | (cur_temp > avg_temp + 2 * std_temp)

cur_temp = get_current_temperature('Beijing')
print("Is anomaly:", is_anomaly(cur_temp, 'Beijing', 'winter'))
```

```python
cur_temp = get_current_temperature('Moscow')
print("Is anomaly:", is_anomaly(cur_temp, 'Moscow', 'winter'))
```

```python
cur_temp = get_current_temperature('New York')
print("Is anomaly:", is_anomaly(cur_temp, 'New York', 'winter'))
```

```python
cur_temp = get_current_temperature('Berlin')
print("Is anomaly:", is_anomaly(cur_temp, 'Berlin', 'winter'))
```

```python
async def get_current_temperature_async(city):
    params = {
        "q": city,
        "appid": API_KEY,
        "units": "metric",
    }
    async with aiohttp.ClientSession() as session:
        async with session.get(WEATHER_URL, params=params) as response:
            data = await response.json()
            return data["main"]["temp"]
```

```python
cities = ['Berlin', "Moscow", "New York", "Beijing"]

tasks = [get_current_temperature_async(city) for city in cities]
results = await asyncio.gather(*tasks)
for city, temp in zip(cities, results):
  print(city, ":", is_anomaly(temp, city, 'winter'))
```
